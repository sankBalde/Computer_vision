# Multidimensional projections in Python

# Iris Dataset
You can check what is the Iris Dataset in here: https://en.wikipedia.org/wiki/Iris_flower_data_set

First, show the dataset with the first two dimensions

```python
import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn import datasets
from sklearn.decomposition import PCA

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

# PCA
We apply PCA and show the first two principal components

```python
# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

# Scaling UMAP
UMAP is superior to t-SNE in that it has a global scale optimization as well as a local-scale optimization.
Therefore, for a large dataset, it can learn the manifold on a portion of the data and project the remaining part of the data very quickly.

```python
from sklearn.model_selection import train_test_split

# Split the digit dataset into a train set and a test set
X_train, X_test, y_train, y_test = train_test_split(digits.data,
                                                    digits.target,
                                                    stratify=digits.target,
                                                    random_state=42)
```

We will use UMAP on the train dataset

```python
trans = umap.UMAP(n_neighbors=5, random_state=42).fit(X_train)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8]
%matplotlib notebook
plt.scatter(trans.embedding_[:, 0], trans.embedding_[:, 1], s= 5, c=y_train, cmap='Spectral')
plt.title('Embedding of the training set by UMAP', fontsize=24)
```

Use the trained model to transform the remaining part of the dataset, the test part:

```python
%time test_embedding = trans.transform(X_test)
```

```python
plt.scatter(test_embedding[:, 0], test_embedding[:, 1], s= 5, c=y_test, cmap='Spectral')
plt.title('Embedding of the test set by UMAP', fontsize=24);
```

Project and visualise this data using t-SNE and UMAP

```python
import json
import pandas as pd
import numpy as np
import umap.umap_ as umap
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import requests

url = "https://github.com/PAIR-code/understanding-umap/raw/master/raw_data/mammoth_umap.json"
response = requests.get(url)

# Enregistre le fichier localement
with open("mammoth_umap.json", "wb") as file:
    file.write(response.content)

print("Fichier tlcharg avec succs.")

data = []
# Ouvre le fichier JSON en mode lecture
with open("mammoth_umap.json", "r") as file:
    data = json.load(file)

projections = data["projections"]
all_projections = []
projection_labels = []
# Appliquer t-SNE  chaque projection disponible
for projection_key in data['projections'].keys():
    projections = data['projections'][projection_key]

    # Convertir les projections en DataFrame
    df_projections = pd.DataFrame(projections, columns=['x', 'y'])
    # Appliquer UMAP
    umap_model = umap.UMAP(n_components=2, random_state=42)
    projections_umap = umap_model.fit_transform(df_projections)
    
    # Ajouter les projections UMAP et les labels  leurs listes respectives
    all_projections.append(projections_umap)
    projection_labels.extend([projection_key] * len(projections_umap))

# Convertir les listes en un DataFrame pour visualisation
df_umap_projections = pd.DataFrame(np.vstack(all_projections), columns=['UMAP1', 'UMAP2'])
df_umap_projections['Label'] = projection_labels
```

```python
# Visualiser les projections UMAP
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_umap_projections['UMAP1'], df_umap_projections['UMAP2'], c=df_umap_projections['Label'].astype('category').cat.codes, cmap='Spectral', s=5)
plt.colorbar(scatter, label='Projection Key')
plt.title('UMAP Projection of All Keys')
plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.show()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
import requests

# Charger les donnes depuis le fichier JSON
with open("mammoth_tsne.json", "r") as file:
    data = json.load(file)

# Crer une liste pour stocker les rsultats des t-SNE de chaque projection
all_projections = []
projection_labels = []

# Appliquer t-SNE  chaque projection disponible
for projection_key in data['projections'].keys():
    projections = data['projections'][projection_key]

    # Convertir les projections en DataFrame
    df_projections = pd.DataFrame(projections, columns=['x', 'y'])

    # Appliquer t-SNE
    tsne = TSNE(n_components=2, random_state=42)
    X_tsne = tsne.fit_transform(df_projections.values)

    # Ajouter les rsultats de t-SNE  une liste, ainsi que les labels des projections
    all_projections.append(X_tsne)
    projection_labels.extend([projection_key] * len(X_tsne))  # Garder une trace des labels

# Convertir les rsultats en un seul tableau pour la visualisation
all_projections_combined = pd.DataFrame(np.vstack(all_projections), columns=['tsne_1', 'tsne_2'])
all_projections_combined['label'] = projection_labels

# Visualiser toutes les projections sur un mme graphique avec des couleurs diffrentes
plt.figure(figsize=(10, 8))

# Utiliser un colormap personnalis ou `tab10` pour un ensemble de couleurs distinctes
colors = plt.cm.get_cmap('tab10', len(data['projections']))

for i, projection_key in enumerate(data['projections'].keys()):
    subset = all_projections_combined[all_projections_combined['label'] == projection_key]
    plt.scatter(subset['tsne_1'], subset['tsne_2'], label=projection_key, color=colors(i), s=20)

# Ajouter une lgende pour distinguer les projections
plt.legend(loc='best')
plt.title('Projection de toutes les projections avec t-SNE')
plt.show()
```
